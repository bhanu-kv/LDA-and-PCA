```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("/content/modified_dataset_1 - dataset_1.csv.csv")
df.index = df.index + 1
```

```python
target = pd.DataFrame(df, columns=["Target"])
df = df.drop('Target', axis=1)
```

```python
df
```

```python
target
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

dataset = pca.fit_transform(df)

dataset = pd.DataFrame(dataset, columns=['PC1','PC2']) 
```

```python
dataset.index = dataset.index + 1
dataset = dataset.merge(target, how='outer', left_index=True, right_index=True)
dataset
```

```python
sns.scatterplot(x="PC1", y="PC2", hue='Target', data=dataset)
```
